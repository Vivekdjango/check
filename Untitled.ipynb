```python
def lat_lon_find(df):
    try:
        speed = 0.33
        t1 = df['date_time'][df['Flag']=='KN'].reset_index(drop=True)[0]
        t2 = df['date_time'][df['Flag']=='AM'].reset_index(drop=True)[0]

        if t1>t2:
            time_diff = (t1-t2).seconds//60
        else:
            time_diff = (t2-t1).seconds//60   
    #     print(time_diff)
        distance = speed*time_diff
    #     print(distance)
        merch = df['merchant name'][df['Flag']=='AM'].reset_index(drop=True)[0]
        lat_lon = df[['lat','long']][df['Flag']=='KN'].values
        amb_merch_loc = amb_df[['Latitude','Longitude']][amb_df['Name']==merch.lower()].values
        dist_list=[i if ((geodesic(j, lat_lon).km)<=distance) else None for i,j in enumerate(amb_merch_loc)]

        n_dist_list = [i for i in dist_list if i!=None]

        if len(n_dist_list)==1:
            df['lat'][df['Flag']=='AM'],df['long'][df['Flag']=='AM'] =amb_merch_loc[n_dist_list[0]][0],amb_merch_loc[n_dist_list[0]][1]
            df['new_flag'] = 'KN'
            return df

        else:
            df['lat'][df['Flag']=='AM'],df['long'][df['Flag']=='AM'] = np.NaN,np.NaN
            df['new_flag'] = df['Flag']
            return df
    except:
        print(df)
```

```python
fin_df = txn_df.groupby(['card no','date']).apply(lambda x:lat_lon_find(x)).reset_index(drop=True)
```
